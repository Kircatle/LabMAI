```python
import matplotlib.pyplot as plt
import scipy.integrate
import numpy as np
import math
import scipy.optimize as opt


class color:
    def __init__(self):
        self.colors =['b', 'g', 'r', 'c', 'm', 'y', 'k', 'w']
        self.i=0
    
    def next(self):
        if self.i<len(self.colors)-1:
            self.i+=1
            return(self.colors[self.i-1])
        else:
            self.i=0
            return(self.colors[len(self.colors)-1])


def func(args):
    try:
        return math.sqrt(2*args[0]-args[1]**2+2*np.log(np.abs(args[1]-1)))
    except:
        return 0 


def main():
    cl = color()
    C=-0.29
    c = []
    while C < 6:
        x= np.arange(-5,5.01,0.01)
        y1 = [float(func([C,x[i]])) for i in range(len(x))]
        y2 = [float(-func([C,x[i]])) for i in range(len(x))]
        lc = cl.next()
        plt.plot(x,y1, color = lc)
        plt.plot(x,y2, color = lc)
        C+=0.5
        c.append(C)
    ax = plt.gca()
    ax.spines['left'].set_position('center')
    ax.spines['bottom'].set_position('center')
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    plt.show()

    
    def sol(C,a,b):
        f  = lambda x : (x**2)/2-math.log(math.fabs(x-1))-C
        return opt.ridder(f,a,b)
    
    def func_int(x):
        return 2/np.sqrt(2*C-x**2+2*np.log(np.abs(x-1)))
    
    y = []
    x = []
    for C in c:
        a = sol(C,-2-C,-(1e-15))
        b = sol(C,0,1-(1e-15))
        x.append(a)
        y.append(scipy.integrate.quad(func_int, a, b)[0])
    fig, ax = plt.subplots()
    ax.plot(x, y)

if __name__ == "__main__":
    main()
```
